# Task 3: Cross-Correlations

## 3.1) Raw cross-correlation histogram
The cross-correlation histogram (CCH) is a standard tool to analyse correlations between two spike trains. The CCH between a reference spike train $x$ and a target spike train $y$ estimates, at each time lag $\tau$, the tendency of $y$ to spike a time $\tau$ after a spike in $x$.
* Analyse simultaneously recorded spike trains ([Data12, Data13], [Data14, Data15], [Data20, Data21], [Data21, Data22], [Data23, Data24]) for possible temporal correlations by use of the CCH (see `xcorr.cch()`).
* Describe the features of the CCHs, and what they mean. What are the differences between the CCHs of the data sets? Are all pairs correlated?

```python
# Load the data
filename1 = 'Data12'
filename2 = 'Data13'
sts1 = misc.load('./data/%s.npy' %filename1)['st']
sts2 = misc.load('./data/%s.npy' %filename2)['st']

# Compute the CCH
w = 1*pq.ms
maxlag = 50*pq.ms
CCH = xcorr.cch(sts1, sts2, w, lag=maxlag, clip=True)

# Plot the CCH
plt.figure(figsize=(12,6))
plt.subplot(2,2,1)
misc.raster(sts1, xlabel='time', ylabel='trial id')
plt.title(filename1)
plt.subplot(2,2,2)
misc.raster(sts2, xlabel='time', ylabel='trial id')
plt.title(filename2)
plt.subplot(3,1,3)
plt.plot(CCH.times.magnitude, CCH.magnitude, 'k-', label='raw CCH')
plt.title('CCH (%s-%s)' %(filename1, filename2))
plt.xlabel('time lag (%s)' %(CCH.times.units.__str__().split(' ')[-1]))
plt.ylabel('counts')
plt.legend()
```

## 3.2) Correlations above chance level?

Estimating the statistical significance of observed correlations is critical to assess rigorously whether two (or more) spike trains are really correlated and the null hypothesis of independence can be rejected. Normally, statistical significance of spike train correlation cannot be assessed analytically. A powerful approach to estimate significance relies on the generation of surrogate data: a surrogate is created by destroying the feature of interest (here, precise correlation) while keeping other relevant features (e.g. firing rates or ISI distribution). 

* Generate several (e.g. 20) surrogate data by spike dithering (see `surr.dithering()`) and compute the CCH between them 
* Compare the surrogate CCH with the CCH of the original data: is the latter significant?
* Play with the dither parameter: what is an optimal choice? Why? (Compare to the firing rates of the surrogate data...) 

```python
# For each data set, generate n_surr surrogates of each trial via spike dithering
dither = 5*pq.ms
n_surr = 20
surr1_dith = [surr.spike_dithering(st, dither, n=n_surr) for st in sts1]
surr2_dith = [surr.spike_dithering(st, dither, n=n_surr) for st in sts2]

# Compute n_surr CCHs, one per pair of surrogate data sets
surrCCH_dith = numpy.zeros((n_surr, len(CCH)))    # Initialize the surrogate CCHs
for surr_id in xrange(n_surr):                   # For each surrogate id j
    x = [st[surr_id] for st in surr1_dith]       # take, for each trial of the 1st and 
    y = [st[surr_id] for st in surr2_dith]       # 2nd data set, its j-th surrogate,
    surrCCH_dith[surr_id, :] = xcorr.cch(         # compute the avg. CCH among them,
        x, y, w, lag=maxlag, clip=True).magnitude # and save it in surrCCH[j]

# Compute the mean CCH, and the threshold for significance
CCHmean = surrCCH_dith.mean(axis=0)
CCHthreshold = CCHmean + 3 * surrCCH_dith.std(axis=0)

# Plot the original data, their CCH, the average from surrogates and the significance threshold
plt.figure(figsize=(12,5))
plt.plot(CCH.times.magnitude, CCH.magnitude, 'k-', label='row CCH')
plt.plot(CCH.times.magnitude, CCHmean, 'g--', lw=2, label='mean surr. CCH')
plt.plot(CCH.times.magnitude, CCHthreshold, 'r--', lw=2, label='significance threshold')
plt.title('CCH (%s-%s) and significance (spike dithering)' %(filename1, filename2))
plt.xlabel('delay (%s)' %(CCH.times.units.__str__().split(' ')[-1]))
plt.ylabel('counts')
plt.legend()
```

## 3.3) Comparison of surrogate types

Spike dithering is not the only way to generate surrogate data. Many other methods exist, which destroy/keep different features of the original data (for a comparison study see Louis et al., 2010).

Among these methods, a simple approach is *spike time randomisation*, which keeps the number of spikes (and therefore the average firing rate) of the original spike trains, but places each spike randomly into the considered time period.

* Generate surrogates of the same spike trains as before by spike time randomisation (see `surr.spike_time_rand()`; work e.g. on `[Data14, Data15]` and `[Data23, Data24]`) 
* Compute the CCH obtained by such surrogate data, and plot it. How is differs from the case of surrogates by dithering?
* Discuss which features does each method preserve, and pros and cons of both 
* On which data sets do they work fine, and on which don't they? Why?

```python
# Generate surrogates by spike time randomisation (str)
surr1_str = [surr.spike_time_rand(st, n=n_surr) for st in sts1]
surr2_str = [surr.spike_time_rand(st, n=n_surr) for st in sts2]

# Compute n_surr CCHs, one per pair of surrogate data sets
surrCCH_str = numpy.zeros((n_surr, len(CCH)))      # Initialize the surrogate CCHs
for surr_id in xrange(n_surr):                    # For each surrogate id j
    x = [st[surr_id] for st in surr1_str]         # take, for each trial of the 1st and in the
    y = [st[surr_id] for st in surr2_str]         # 2nd data set, their j-th surrogate, 
    surrCCH_str[surr_id, :] = xcorr.cch(          # compute their CCH 
        x, y, w, lag=maxlag, clip=True).magnitude # and save it in surrCCH[j]

# Compute the mean CCH, and the threshold for significance
CCHmean_str = surrCCH_str.mean(axis=0)
CCHthreshold_str = CCHmean_str + 3 * surrCCH_str.std(axis=0)

# Plot the original CCH, the average from surrogates and the significance threshold
plt.figure(figsize=(12,5))
plt.plot(CCH.times.magnitude, CCH.magnitude, 'k-', label='row CCH')
plt.plot(CCH.times.magnitude, CCHmean_str, 'g--', lw=2, label='mean surr. CCH')
plt.plot(CCH.times.magnitude, CCHthreshold_str, 'r--', lw=2, label='significance threshold')
plt.title('CCH (%s-%s)  and significance (spike time randomisation)' %(filename1, filename2))
plt.xlabel('delay (%s)' %(CCH.times.units.__str__().split(' ')[-1]))
plt.ylabel('counts')
plt.legend()
```

# Task 4: Higher-order correlations

## 4.1) Compound Poisson Process (CPP)

The Compund Poisson Process (CPP) is a versatile model of parallel spike trains correlated by a well-defined correlation structure. Here, we will use the CPP to generate various types of correlated data, that we will later analyse.

* Generate correlated parallel spike trains (e.g. 100) by means of the CPP (see `stocmod.cpp()`). 
* Plot the data: are correlations visible in the raster plot by eye? 
* Plot also the population histogram and the histogram of its values (the *complexity distribution*): is the correlation visible there? Try different firing rates and amplitude distributions.
* Discuss the role of the amplitude distribution in determining the correlation structure among the parallel spike trains.
* Discuss the relation between the amplitude distribution and the complexity distribution.

```python
# Specify number of parallel spike trains, their rate, the simulation time
n = 100      
r = 20*pq.Hz 
T = 10*pq.s

# Specify the amplitude distribution of the model
A = numpy.zeros(n+1) 
A[[1,10]] = .98, .02

# Generate the CPP data
cpp = stocmod.cpp(A, T, r)

# Compute the population histogram and the complexity distribution
binsize = 10*pq.ms
pophist = stats.peth(cpp, w=binsize)
compl_distr, edges = numpy.histogram(pophist, bins=xrange(n+2))
compl_distr = compl_distr/float(sum(compl_distr))

# Plot the results
fig = plt.figure(figsize=(12,5))
fig.subplots_adjust(top=.92, bottom=.05, hspace=.5, wspace=.2)
misc.add_raster(fig, (2,2,1), cpp, ms=1, xlabel='time', ylabel='trial id')
misc.add_hist(fig, (2,2,3), pophist.times, pophist, pophist.sampling_period, xlabel='time', ylabel='counts')
ylim = plt.gca().get_ylim()
plt.subplot(1,2,2)
plt.bar(edges[:-1], A, color='r', label='amplitude distrib.')
plt.plot(edges[:-1], compl_distr, label='complexity distrib.')
plt.ylim([0, 0.25])
plt.xlabel('probability', size=12)
plt.ylabel('complexity', size=12)
plt.suptitle('CPP', size=14)
plt.legend()
```

## 4.2) Investigating correlations by the complexity distribution

The complexity distribution carries information about the probability of observing any number of synchronous spikes in the data. Such synchrony can be due to chance, or to correlations in the spiking activity. It will most of the time be a sum of the two.

Generating control, independent data and comparing their complexity distribution to that of the original data can be an effective way to assess the presence of higher-order correlations (see Grün et al (2008)).

* Generate control data in terms of an independent CPP with same marginal statistics (number of neurons, firing rate)
* Compare the two amplitude distributions (e.g. by plotting their difference): are correlations visible in the correlated data?

```python
# Generate independent CPP process
A_ind = numpy.zeros(n+1)
A_ind[1] = 1.
cpp_ind = stocmod.cpp(A_ind, T, r)

# Compute the population histogram and the complexity distribution
pophist_ind = stats.peth(cpp_ind, w=binsize)
compl_distr_ind, edges = numpy.histogram(pophist_ind, bins=xrange(n+2))
compl_distr_ind = compl_distr_ind / float(sum(compl_distr_ind))

# Plot the difference of the complexity distributions of the correlated and independent CPP
plt.figure()
plt.subplot(1,2,1)
plt.plot(edges[:-1], compl_distr, color='blue', label="corr'd")
plt.plot(edges[:-1], compl_distr_ind, color='red', label="indep")
plt.xlabel('complexity')
plt.ylabel('probability')
plt.legend()

plt.subplot(1,2,2)
plt.plot(edges[:-1], compl_distr-compl_distr_ind)
plt.xlabel('complexity')
plt.ylabel('probability diff.')
plt.show()
```

## 4.3) CuBIC
The CuBIC (Cumulant Based Inference of Correlation) analysis formalises the intuition that, at population level, the complexity distribution carries the information about the correlation in the data (See Staude et al (2010a,b); Grün and Rotter (2010)). 

Memo from the lecture: CuBIC analyses the cumulants of the complexity distribution and runs statistical tests on those in an iterative fashion. At each step, a certain maximum correlation order $\xi$ is accepted under the null hypothesis, while the alternative hypothesis allows for correlations of order higher than $\xi$. The iteration stops when the null hypothesis cannot be rejected.

* Analyse the population correlation of the CPP data by means of CuBIC (see `cubic.cubic()`)
* Is the order of correlation inferred correctly? Or is there a bias with respect to the real order of correlation? Why?
* Which important constraints do the cumulants of the complexity distribution have to satisfy in order for CuBIC to solve the equations? Check them!
* Play with the bin size of the population histogram used for the analysis: do the results change?

```python
# Analyse the correlations in the data via CuBIC
w = 5*pq.ms
alpha = .99
pophist = stats.peth(cpp, w=w)
xi, p, k = cubic.cubic(pophist, alpha=alpha)

print 'cumulants: ', k
xi_true = numpy.where([A>0])[-1][-1]

ax = plt.subplot(111)
ax.bar(left=numpy.arange(.5, len(p)+.5), height=p, color='darkgreen')
ax.set_xlim(.5, len(p)+.5)
ax.plot(ax.get_xlim(), [alpha, alpha], 'r--', lw=2)
ax.text(.1, .9, r'$\hat\xi=%d$ $(\xi_{true}=%d)$' %(xi, xi_true), 
        transform=ax.transAxes, size=14)
plt.show()
```

## 4.5) Non-stationary CuBIC
When estimating the order of population correlation, standard CuBIC
assumes that firing rates are stationary over time. Non-stationary firing
rates might bias the significance of observed correlations. Analytical 
extensions have been derived to account for some rate distributions, 
namely cosine-, uniform- and gamma-distributed rates.

* Generate an independent CPP process (amplitude distribution with non-zero value in 1 only) with cosine rate profile (see `stocmod.cpp_cos()`) 
* Analyse it with standard and cosine CuBIC. What do you observe?



```python
# Simulate a CPP with sinusoidal firing rates and no correlations
T = 10*pq.s # simulation time
N = 100     # number of spike trains in the CPP

A = numpy.zeros(N)   # initialize amplitude distribution
A[1] = 1             # define strength of different correlation orders
xi_true = numpy.where([A>0])[-1][-1] 

a = 6*pq.Hz
b = 10*pq.Hz
w = 2*pq.Hz
cpp_cos = stocmod.cpp_cos(A, T, a=a, b=b, w=w, phi=0)

# Analyse the data with CuBIC, assuming stationary (default) and cosine rate profile
w = 5*pq.ms
pophist = stats.peth(cpp_cos, w=w)

alpha = 0.05
xi1, p1, k1 = cubic.cubic(pophist, alpha=alpha)
xi2, p2, k2 = cubic.cubic(pophist, alpha=alpha, rate_distr = 'cos')

# Plots
fig = plt.figure(figsize=(16,5))
fig.subplots_adjust(top=.92, bottom=.05, hspace=.5, wspace=.2)
misc.add_raster(fig, (2,1,1), cpp_cos, xlabel='time', ylabel='trial id')

ax = fig.add_subplot(223)
ax.bar(left=numpy.arange(.5, len(p1)+.5), height=p1, color='darkgreen')
ax.set_xlim(.5, len(p1)+.5)
ax.set_ylim((0, .2))
ax.plot(ax.get_xlim(), [alpha, alpha], 'r--', lw=2)
ax.text(.1, .8, r'$\hat\xi=%d$ $(\xi_{true}=%d)$' %(xi1, xi_true), 
        transform=ax.transAxes, size=14)
ax.set_title('CuBIC - stationary')

ax = fig.add_subplot(224)
ax.bar(left=numpy.arange(.5, len(p2)+.5), height=p2, color='darkgreen')
ax.set_xlim(.5, len(p2)+.5)
ax.set_ylim((0, .2))
ax.plot(ax.get_xlim(), [alpha, alpha], 'r--', lw=2)
ax.text(.8, .8, r'$\hat\xi=%d$ $(\xi_{true}=%d)$' %(xi2, xi_true), 
        transform=ax.transAxes, size=14)
ax.set_title('CuBIC - cosine')
```
