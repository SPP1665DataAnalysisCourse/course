## 1.2) Peri-event-time-histogram

The spiking activity in response to a stimulus can be estimated by repeatedly presenting the stimulus, and averaging the response across trials. Averaging the spike counts in short time bins across trials aligned to the same trigger (e.g. the stimulus presentation) yields the so-called *peri-event-time histogram* (PETH). 

* Estimate the firing rate profile averaging across the spike counts at each bin across trials (see `stats.peth()`). 
* Use different bin widths [e.g: `stats.peth(sts, w=10*pq.ms)`]. What is a good choice? Why?
* Vary, for a fixed bin size, the number of trials considered [e.g. `stats.peth(sts[:10], w=10*pq.ms)`, or `stats.peth(sts[::2], w=10\*pq.ms)`]. How do the results change?

```python
# Compute PETH
w = 20*pq.ms
peth = stats.peth(sts, w=w, output='rate').rescale('Hz')

# Plot results
misc.hist(left=peth.times, height=peth, width=peth.sampling_period, ylabel='firing rate')
```

## 1.3*) Rate estimation via kernel convolution

Estimating the firing rate of a neuron via the PETH requires to have multiple trials for that neuron's acitivity, and requires the statistical properties of the activity to be stationary across trials. Often these two assumptions are not met, either because multiple trials are not available, or because the statistics change across trials (see e.g. `Data2`, `Data3`).

A possible method to draw an estimate of the firing rate profile from a single spike train is to smooth (mathematically, convolve) each spike with a continuos kernel (see `rate_est.instaneous_rate()`).

Consider, in the data you are working with, only a single trial (e.g. `sts[0]`). Create a smooth estimate of the firing rate via kernel convolution, and plot it. (You can use the default "TRI" (triangular) kernel).
* Is this approximation consistent with the PETH estimate? Why?
* How does the result depend on the kernel width? Try different values. What is an optimal kernel width?

```python
rate_inst = rate_est.instantaneous_rate(spiketrain=sts[0], resolution=0.1*pq.ms, form='TRI', sigma=0.3*pq.s, trim=False)
plt.plot(rate_inst.times.rescale('ms').magnitude, rate_inst.rescale('Hz').magnitude)
plt.title('Instantaneous firing rate')
plt.xlabel('Time (ms)')
plt.ylabel('Rate (Hz)')
```

## 1.3) Cross-trial variability and Fano Factor

An implicit assumption when averaging a quantity (here, the spike count) across trials is that such quantity is stationary (i.e. preserves its statistics) across trials. Deviations from this asssumption might lead to misleading interpretation of the results.

* Display the number of spikes in each trial of the previous data. Is the spike count stationary across trials?

A common measure of cross-trial spike count variability is the Fano Factor (FF), defined as the normalised variance of spike counts across trials:
        
             FF = var(spike counts) / mean(spike counts)
            
* Compute the FF of trial spike counts (see `stats.fanofactor()`)
* Discuss whether the PETH is a reliable estimate of the firing rate for the data under consideration.

```python
# Plot the spike counts per trial
plt.plot(range(1,len(sts)+1), [len(st) for st in sts], '.-', ms=10)
plt.xlabel('trial id')
plt.ylabel('spike count')

# Compute the Fano Factor of the spike counts
print 'FF =', stats.fanofactor(sts)
```

# 3) ISI variability
## 3.1) Coefficient of Variation (CV)
The ISI variability is an interesting measure of regularity of the spiking process. ISI varaibility is normally measured by the Coefficient of Variation (CV), defined by

        CV(ISIs) = sd(ISI) / mean(ISI) ,
        
For instance, for a stationary Poisson process, ISIs are exponentially distributed, and their theoretical CV is 1. CV < 1 indicates higher regularity in the spiking activity (CV = 0 being the extreme case of a clock process), while CV > 1 indicates higher irregularity. 

* Estimate the CV of the ISIs for the data you are considering (see `stats.cv()`). Is the CV high or low? What does that mean?
* Do the data look "Poissonian"? Fit the estimated ISI density to an exponential probability density (see `misc.fit_to_exp()`). You can estimate the goodness of fit by computing the correlation coefficient between the real and the fit ISI distribution (see `numpy.corrcoef()`)

```python
# Compute the CV
cv = stats.cv(sts)


# Fit ISI distribution to exponential 
r, pdf_fit = misc.fit_to_exp(ISIpdf.times, ISIpdf)

print 
# Plot the ISI histogram and the fit
plt.figure()
misc.hist(ISIpdf.times, ISIpdf, width=w, xlabel='ISI', ylabel='probability')
plt.plot(ISIpdf.times.magnitude, pdf_fit.magnitude, '--r', lw=2)
plt.text(.95, .8, r'$\hat\lambda$=%.2f' % (r.rescale('Hz')[0]), transform=plt.gca().transAxes, size=16, ha='right')
plt.text(.95, .6, r'$\hat{CV}$=%.2f' % cv, transform=plt.gca().transAxes, size=16, ha='right')
plt.show()
```

## 5) Renewalty of the Spike train: FF vs CV$^2$
From the theory of point processes we know that renewal processes which are stationary in time satisfy the relation

                                    FF=CV$^2$
Thus, if the relation is not satisfied the process under consideration cannot be both renewal and stationary. In particular, if we know that it is stationary we can draw the conclusion that it is not renewal.
* Load the following data, each corresponding to the spiking activity of one (real or simulated) unit: `Data10`, `Data12`, `Data26`, `Data27`
* Compute the FF and the CV$^2$ of each data set
* For each data set, plot the FF vs CV$^2$; as a term of comparison, plot also the identity line
